## Applying Functions to Cells, Columns and Rows

To apply functions to each column, use apply():

```python
data.apply(np.max)
```

The apply method can also be used to apply a function to each row. To do this, specify axis=1. Lambda functions are very convenient in such scenarios. For example, if we need to select all states starting with W

```python
data[data['state'].apply(lambda state: state[0] == 'W')].head()
```

The map method can be used to replace values in a column by passing a dictionary of the form {old_value: new_value} as its argument:

```python
d = {'no' : False, 'yes' : True}
data['international plan'] = data['international plan'].map(d)
data.head()
```

The same thing can be done with the replace method:

```python
d = {'no' : False, 'yes' : True}
data = data.replace({'voice mail plan': 'd'})
data.head()
```

## Ploting

```python
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
```

```python
sns.countplot(x='international plan', hue='churn', data=data);
```

```python
pd.crosstab(data['churn'], data['customer service calls'], margins=True)
```

```python
sns.countplot(x='customer service calls', hue='churn', data=data);
```

Although it's not so obvious from the summary table, it's easy to see from the above plot that the churn rate increases sharply from 4 customer service calls and above.

Now let's add a binary feature to our DataFrame â€“ Customer service calls > 3

```python
data['many_service_calls'] = (data['customer service calls'] > 3).astype('int')
pd.crosstab(data['many_service_calls'], data['churn'], margins=True)
```

```python
sns.countplot(x='many_service_calls', hue='churn', data=data);
```

Let's construct another contingency table that relates Churn with both International plan and freshly created Many_service_calls.

```python
pd.crosstab(data['many_service_calls'] & data['international plan'] , data['churn'])
```
